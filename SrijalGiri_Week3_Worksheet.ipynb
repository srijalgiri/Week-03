```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dict={'Name':['Ram','Shyam','Mohan','Peter','Hari'],
          'Age':[22,18,17,19,23],
           'Score':[50,60,80,77,88]}
df=pd.DataFrame(data_dict)
df.sort_values(by='Score')
```

```python
df_index=df.set_index('Name')
df_index
```

```python
df.reset_index(drop=True)
```

```python
#Subsetting using iloc and loc
df.iloc[0:2,0:2]
```

```python
#Subsetting using loc
df.loc[df["Age"]>18,['Name','Age']]
```

```python
df.loc[(df['Age']>17) & (df['Score']>70),['Name','Score']]
```

#Grouping and Aggregation

```python
df.groupby('Age')['Score'].mean()
```

```python
df.groupby('Age')['Score'].count()
```

```python
df.groupby('Age').agg({'Score':['mean','sum','count']})
```

```python
df.groupby('Age').describe()
```

#Plotting

```python
df.plot(x='Score',y='Age',xlabel='Score',ylabel='Age',kind='bar',title="Age vs Score,",color='crimson')
plt.show()
```
